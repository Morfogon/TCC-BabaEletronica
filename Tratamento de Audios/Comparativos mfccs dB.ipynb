```python
x1, sr1 = librosa.load('Dataset_Microfone/Choro/ChoroMic1.wav')
mfccs1 = librosa.feature.mfcc(y=x1, sr=sr1, n_mfcc=26)
features1 = np.mean(mfccs1, axis=1)

x2, sr2 = librosa.load('Crying_LimpoV2/4-167077-A_resampled.wav')
mfccs2 = librosa.feature.mfcc(y=x2, sr=sr2, n_mfcc=26)
features2 = np.mean(mfccs2, axis=1)
```

```python
fig = plt.figure(figsize=(17, 4))
plt.plot(features1, label='ChoroMic1.wav',color='red')
plt.xlabel('Numero de MFCCs')
plt.ylabel('Valores de MFCCs')
plt.title('MLPClassifier com 100 camadas internas')
plt.plot(features2, label='4-167077-A_resampled.wav',color='green')
plt.legend()
plt.show()
```
