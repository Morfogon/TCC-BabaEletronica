```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Carrega os arquivos de áudio
audio1N = 'Louise_01.m4a_10.wav'
audio2N = 'Som_Do_BebeV3.wav'
audio1, sr1 = librosa.load(audio1N)
audio2, sr2 = librosa.load(audio2N)
print(f'taxa de amostragem de {audio1N} é {sr1}')
print(f'taxa de amostragem de {audio2N} é {sr2}')
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Gera o espectrograma para o primeiro áudio
spectrogram1 = librosa.stft(audio1, n_fft=2048, hop_length=512)
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.amplitude_to_db(abs(spectrogram1)), sr=sr1, x_axis='tempo', y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title(f'Espectrograma - {audio1N}')
plt.tight_layout()
plt.show()

# Gera o espectrograma para o segundo áudio
spectrogram2 = librosa.stft(audio2, n_fft=2048, hop_length=512)
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.amplitude_to_db(abs(spectrogram2)), sr=sr2, x_axis='tempo', y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title(f'Espectrograma - {audio2N}')
plt.tight_layout()
plt.show()

# Gera o gráfico de frequência para o primeiro áudio
freq1 = librosa.fft_frequencies(sr=sr1, n_fft=2048)
mag1 = librosa.amplitude_to_db(abs(spectrogram1))
plt.figure(figsize=(10, 4))
plt.plot(freq1, mag1[:,0], color='black')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title(f'Gráfico de Frequência - {audio1N}')
plt.tight_layout()
plt.show()

# Gera o gráfico de frequência para o segundo áudio
freq2 = librosa.fft_frequencies(sr=sr2, n_fft=2048)
mag2 = librosa.amplitude_to_db(abs(spectrogram2))
plt.figure(figsize=(10, 4))
plt.plot(freq2, mag2[:,0], color='black')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title(f'Gráfico de Frequência - {audio2N}')
plt.tight_layout()
plt.show()
```
